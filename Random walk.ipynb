```python
import plotly.graph_objects as go
import random

def random_walk_3d(num_steps):
    # Initialize starting point
    x = [0]
    y = [0]
    z = [0]

    # Simulate random walk
    for _ in range(num_steps):
        # Generate random direction
        direction = random.choice(['x', 'y', 'z'])
        
        # Update coordinates based on direction
        if direction == 'x':
            x.append(x[-1] + random.choice([-1, 1]))
            y.append(y[-1])
            z.append(z[-1])
        elif direction == 'y':
            x.append(x[-1])
            y.append(y[-1] + random.choice([-1, 1]))
            z.append(z[-1])
        else:
            x.append(x[-1])
            y.append(y[-1])
            z.append(z[-1] + random.choice([-1, 1]))

    return x, y, z

# Set the number of steps in the random walk
num_steps = 10000

# Simulate the random walk
x, y, z = random_walk_3d(num_steps)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

# Set plot title and axis labels
fig.update_layout(title='3D Random Walk',
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))

# Show the plot
fig.show()
```
