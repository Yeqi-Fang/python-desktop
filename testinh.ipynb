```python
import numpy as np
import matplotlib.pyplot as plt

def NevilleNewton(x, a, b):
    n = len(a)
    m = np.zeros(n)
    f = np.zeros((n, n))
    for i in range(n):
        r = 1.0
        for j in range(i):
            r *= (x - a[j])
        m[i] = r
    for i in range(n):
        f[i, 0] = b[i]
    for i in range(1, n):
        for j in range(i, n):
            f[j, i] = (f[j, i-1] - f[j-1, i-1]) / (a[j] - a[j-i])
    for i in range(1, n):
        for j in range(i):
            f[j, i] = 0.0
    d = np.sum(f.diagonal() * m)
    return d

n = int(input("请输入插值点个数n："))
a = np.zeros(n)
b = np.zeros(n)
for i in range(n):
    a[i] = float(input("请输入第{}点x坐标：".format(i)))
for i in range(n):
    b[i] = float(input("请输入第{}点y坐标：".format(i)))
x = float(input("请输入要求点的x坐标："))
print(NevilleNewton(x, a, b))
fig, ax = plt.subplots()
ax.scatter(a, b, color='red', label='Interpolation Points')
x_i = np.linspace(a.min(), x, 1000)
y_i = np.zeros_like(x_i)
for i in range(len(x_i)):
    y_i[i] = NevilleNewton(x_i[i], a, b)
ax.plot(x_i, y_i, color='blue', label='Interpolating Polynomial')
ax.legend()
plt.show()
```
